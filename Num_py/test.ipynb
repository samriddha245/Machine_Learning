Create 2d array and multiply

```python
import numpy as np

arry2d1 = np.array([[4,5,6],[6,7,8],[7,8,9]])
arry2d2 = np.array([[1,2,3],[3,4,5],[6,7,8]])

print(arry2d1 * arry2d2)

```

```python
ary2d1=np.array([[4,5,6],[6,7,8],[7,8,9]])
ary2d2=np.array([[1,2,3],[3,4,5],[6,7,8]])
print(ary2d1@ary2d2)
```
