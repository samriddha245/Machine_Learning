## Mean and mode imputation


```python
gender_mode = student_df['gender'].mode()[0]
```

```python
gender_mode
```

```python
marks_mean=student_df['marks_obtained'].mean()
marks_mean
```

```python
student_df['gender']=student_df['gender'].fillna(gender_mode)
student_df
```

```python
student_df['marks_obtained']=student_df['marks_obtained'].fillna(marks_mean)
student_df
```
